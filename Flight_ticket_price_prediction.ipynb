```python
df.drop('flight',axis=1,inplace=True)
```

```python
df['stops']=df['stops'].map({'zero':0,'one':1,'two_or_more':2})
```

```python
sns.catplot(y='price',x='departure_time',data=df.sort_values('price',ascending=False),kind='boxen',height=6,aspect=3)
plt.show()
```

```python
plt.scatter(x=df.source_city,y=df.price,c=np.random.rand(len(df),3),facecolor='maroon',s=500)
plt.xlabel(" source_city  ")
plt.ylabel(" price ")
plt.title("Price based on Source city")
plt.show()
```

```python
sns.catplot(y='price',x='destination_city',data=df.sort_values('price',ascending=False),kind='boxen',height=6,aspect=3)

plt.title("Price based on Destination city")

plt.show()
```

```python
plt.scatter(x=df.airline,y=df.stops,cmap='tab20')

# provide a range for numbers to colors based on the data plotted in the graph
plt.colorbar()

plt.xlabel("Airline ")
plt.ylabel("stops ")

plt.title("Airline based on stoppies")

plt.show()
```

```python
sns.catplot(y='duration',x='source_city',data=df.sort_values('source_city',ascending=False),kind='box')
plt.title("duration based on source city")

plt.show()
```

```python
# alpha argument is used to make the scatter markers transparent.
# By default it values is 1 i.e. no transparency.

plt.scatter(y=df.price,x=df.airline,color=['darkorange'],s=800,alpha=0.11)

plt.xlabel("Price")
plt.ylabel("Airline")
plt.title("Price based on Airline")

plt.show()
```
